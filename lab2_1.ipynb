**6. Дан список из чисел. Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий
делитель).**

```python
def gcd(first, second):
    """Вернуть НОД целых чисел 'first' и 'second'."""
    mn = min(first, second)
    d = []
    for i in range(1, mn + 1):
      if first % i == 0 and second % i == 0:
        d.append(i)
    return max(d)


def lcm(first, second):
    """Вернуть НОК целых чисел 'first' и 'second'."""
    mx = max(first, second)
    for i in range(mx, first * second + 1):
      if i % first== 0 and i % second == 0:
        return i

def gcd_nums(nums):
    """Вернуть НОД чисел из списка 'nums'. """
    r = nums[0]
    for i in range(1, len(nums)):
      r = gcd(r, nums[i])
    return r



def lcm_nums(nums):
    """Вернуть НОК чисел из списка 'nums'. """
    o = nums[0]
    for i in range(1, len(nums)):
      o = lcm(o, nums[i])
    return(o)

nums = list(map(int,(input()).split()))
print(gcd_nums(nums))
print(lcm_nums(nums))
```

**9. Напишите функции для перевода числа (N∈[2;16]):
1.из 10-й системы счисления в N-ю;
2.из N-й системы счисления в 10-ю.**

```python
LETTERS_EX = {10: "A", 11: "B", 12: "C", 13: "D", 14: "E", 15: "F"}
DIGITS_EX = {"A": 10, "B": 11, "C": 12, "D": 13, "E": 14, "F": 15}


def to_base(number, base):
    """Преобразовать десятичное число 'number' в с.с. 'base'."""
    d = []
    while number > 0 :
      c = number % base
      if c < 10:
        d.append(str(c))
      else:
        d.append(LETTERS_EX[c])
      number = number // base
    d.reverse()
    return ''.join(d)


def from_base(number, base):
    """Преобразовать число в десятичную систему счисления"""
    number_str = str(number)
    values = []

    for position, digit_char in enumerate(number_str[::-1]):
        if digit_char in DIGITS_EX:
            digit_value = DIGITS_EX[digit_char]
        else:
            digit_value = int(digit_char)

        values.append(digit_value * (base ** position))
    return sum(values)

base = int(input("Введите основание: "))
number = int(input("Введите число для перевода: "))
number_2 = int(input("Введите число для перевода в десятичную систему: "))

print(to_base(number, base))
print(from_base(number, 10))
```

```python
def sentence_stats(sentence):
    """Вернуть символьную статистику 'sentence'. Регистр не учитывается."""
    l = sentence.lower()
    sent_low = list(set(l))
    value = []
    for i in sent_low:
      v = 0
      for j in l:
        if j == i:
          v += 1
      value.append(v)
    res = dict(zip(sent_low, value))
    return res

sentence = str(input())
print(sentence_stats(sentence))
```

**11. Используя шифр Цезаря (достаточно только букв русского алфавита, знаки
препинания не изменяются), зашифруйте, а затем расшифруйте введенную строку**

```python
def ceasar(text, shift):
    """Вернуть измененную строку 'text' со сдвигом 'shift'."""
    # Набор кириллических букв
    letters = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    ind = []
    for i in text:
      if i in letters:
        n = letters.index(i)
        k = n + shift
        ind.append(letters[k])
      elif i.lower() in letters:
        nr =  letters.index(i.lower())
        nm = nr + shift
        ind.append(letters[nm].upper())
    return "".join(ind)


text = str(input())
shift = int(input())

encoded = ceasar(text, shift)
decoded = ceasar(encoded,-shift )
print("Зашифрованная строка:", encoded)
print("Расшифрованная строка:", decoded)
```

**13. Дан список с результатами голосования на выборах в виде:
[1, 3, 2, 2, 2, 5, -1, ...]
где номер определяет голос за партию из списка:
1.Партия №1.
2.Партия №2.
3.Партия №3.
4.Партия №4.
5.Партия №5.
-1.Испорченный бланк.
Подведите итоги выборов, выведя на экран список партий в соответствии с
убыванием количества полученных голосов и их процентным соотношением:
1.Партия №2 | 1111 | 58.21%
2.Партия №4 | 999 | 38.14%**

```python
votes = [
    2, 3, -1, 2, 5, 1, 1, 4, 1, 1, 1, 3, 1, 3, 5, -1, 5, 2, 5, 5,
    3, 3, 2, 3, 3, 2, 2, 1, 3, 2, 5, 2, 2, 4, 1, 1, 3, 2, 2, 3, 3,
    3, 1, 4, 2, 1, 4, 2, 3, 3, 3, -1, 5, 3, 1, 4, 5, 1, 1, 3, 3,
    3, -1, 5, 3, 3, 2, 5, 1, 1, 5, -1, 1, 2, 2, 3, -1, 4, 2, 5, 4,
    2, -1, 3, 1, 4, 3, 5, 4, 1, 5, 3, 2, 4, 2, 1, 3, 4, 1, 1, 3, 4
]

parties = {
    1: "Партия №1", 2: "Партия №2", 3: "Партия №3",
    4: "Партия №4", 5: "Партия №5", -1: "Испорчено"
}


def parties_votes(parties, votes):
    """Вернуть информацию о голосах 'votes', отданных за партии 'parties'."""
    result = {}
    for vote in votes:
        party_name = parties[vote]
        if party_name in result:
            result[party_name] += 1
        else:
            result[party_name] = 1
    return result


def print_results(votes_for_p):
    """Вывести результаты голосования."""
    # Сортируем по убыванию голосов
    sorted_parties = sorted(votes_for_p.items(), key=lambda x: x[1], reverse=True)

    # Подсчитываем общее количество голосов (без испорченных)
    total_valid = sum(count for party, count in votes_for_p.items() if party != "Испорчено")

    # Выводим результаты
    for i, (party, count) in enumerate(sorted_parties, 1):
        if party != "Испорчено":
            percentage = (count / total_valid) * 100
            print(f"{i}. {party} | {count} | {percentage:.2f}%")
        else:
            print(f"{i}. {party} | {count} | -")


print_results(parties_votes(parties, votes))
```

```python
def pow1(value, power):
    """Вернуть 'value' в степени 'power'. !!!."""
    res = value ** power
    return res


def pow2(value, power):
    """Вернуть 'value' в степени 'power'. Рекурсивный алгоритм.."""
    if power == 0:
      return 1
    else:
      return value * pow2(value, power - 1)



print(pow1(5, 3))
print(pow2(5, 3))
```

**17. Дано натуральное число. Напишите рекурсивные функции для определения:
1.суммы цифр числа;
2.количества цифр в числе.**

```python
def digits_sum(value):
    """Вернуть сумму цифр числа 'value'. Рекурсивная реализация. """
    if value < 10:
      return value
    else:
      return (value % 10) + digits_sum(value // 10)



def digits_count(value):
    """Вернуть количество цифр числа 'value'. Рекурсивная реализация."""
    if value < 10:
      return 1
    else:
      value = value // 10
    return 1 + digits_count(value)


print(digits_sum(12345))
print(digits_count(12345))
```

```python
def digits_sum(value):
    """Вернуть сумму цифр числа 'value'. Рекурсивная реализация. """
    if value < 10:
      return value
    else:
      return (value % 10) + digits_sum(value // 10)

print(digits_sum(12345))
print(digits_sum(987))
```

последовательности Фибоначчи. Последовательность Фибоначчи — это ряд чисел,
в котором первые два числа равны 0 и 1, а каждое последующее число равно сумме
двух предыдущих.**



```python
def fibonacci_rec(n):
    """Найти n-ое число Фибоначчи рекурсивно."""
    if n == 0:
      return 0
    elif n == 1:
      return 1
    else:
      return fibonacci_rec(n - 1) + fibonacci_rec(n - 2)

print(fibonacci_rec(10))
print(fibonacci_rec(5))
```

**20. Создайте рекурсивную функцию, которая переворачивает строку. Например, для
строки "hello" результат должен быть "olleh".**

```python
def reverse_string_rec(s):
    """Перевернуть строку 's' с помощью рекурсии."""
    if  len(s) == 1:
      return s
    else:
      return s[-1] + reverse_string_rec(s[:-1])

print(reverse_string_rec("hello"))
print(reverse_string_rec("python"))
```

**21. Напишите рекурсивную функцию, которая проверяет, является ли строка
палиндромом. Палиндром — это строка, которая читается одинаково в обоих
направлениях (например, "racecar").**

```python
def is_palindrome_rec(s):
    """Проверить, является ли строка 's' палиндромом, рекурсивно."""
    if len(s) <=  1:
      return True
    if s[0] != s[-1]:
      return False
    else:
      return is_palindrome_rec(s[1:-1])


print(is_palindrome_rec("racecar"))
print(is_palindrome_rec("hello"))
print(is_palindrome_rec("madam"))
```

**22. Разработайте рекурсивную функцию для вычисления суммы всех элементов в
списке чисел**

```python
def sum_list_rec(numbers):
    """Вычислить сумму элементов списка 'numbers' рекурсивно"""
    if len(numbers) == 0:
      return 0
    else:
      return numbers[0] + sum_list_rec(numbers[1:])

my_list = [1, 5, 3, 7, 2, 8]
print(sum_list_rec(my_list))
print(sum_list_rec([]))
```

**24. Реализуйте рекурсивную функцию для нахождения наибольшего общего делителя
(НОД) двух чисел, используя алгоритм Евклида**

```python
def gcd_rec(a, b):
    """Найти наибольший общий делитель (НОД) чисел 'a' и 'b' рекурсивно."""
    if a == 0 or b == 0:
      return max(a,b)
    if a % b == 0 or b % a == 0:
      return (min(a,b))
    else:
      if a > b:
       return gcd_rec(a % b, b)
      else:
        return gcd_rec(b % a, a)


print(gcd_rec(0, 18))
print(gcd_rec(101, 10))
```

**25. Напишите рекурсивную функцию, которая преобразует десятичное число в его
двоичное представление в виде строки.**

```python
def to_binary_rec(n):
    """Преобразовать десятичное число 'n' в двоичную строку рекурсивно."""
    if n // 2 == 0:
      return "1"
    else:
      return to_binary_rec(n // 2) + str(n % 2)


print(to_binary_rec(255))
print(to_binary_rec(7))
print(to_binary_rec(0))
```

**26. Напишите рекурсивную функцию, которая "сплющивает" вложенный список, то
есть делает из списка с подсписками один плоский список.**

```python
def flatten_list_rec(nested_list):
    """Преобразовать вложенный список 'nested_list' в плоский рекурсивно."""
    res = []
    if len(nested_list) == 0:
      return res

    first = nested_list[0]
    s = nested_list[1:]
    if isinstance(first, list):
      res.extend(flatten_list_rec(first))
    else:
      res.append(first)
    res.extend(flatten_list_rec(s))
    return res

my_list = [1, [2, 3], 4, [[5, 6], 7]]

print(flatten_list_rec(my_list))
```

**27. Создайте рекурсивную функцию для подсчета количества вхождений
определенного элемента в список**

```python
def count_occurrences_rec(lst, item):
    """Подсчитать, сколько раз 'item' встречается в списке 'lst', рекурсивно."""
    if len(lst) == 0:
      return 0
    else:
      return int(lst[0] == item) + count_occurrences_rec(lst[1:], item)


my_list = [1, 2, 5, 3, 2, 4, 2, 5]
print(count_occurrences_rec(my_list, 2))
print(count_occurrences_rec(my_list, 5))
print(count_occurrences_rec(my_list, 9))
```

**28. Реализуйте рекурсивную функцию digital_root(n), которая находит цифровой
корень числа. Цифровой корень — это однозначное число, полученное путем
итеративного процесса суммирования цифр. Например, для 65536: 6+5+5+3+6=25,
затем 2+5=7. Цифровой корень — 7.**

```python
def digital_root_rec(n):
    """Найти цифровой корень числа 'n' рекурсивно."""
    if n == 0:
      return 0
    else:
      s = n % 10 + digital_root_rec(n // 10)
    if len(str(s)) == 1:
      return s
    else:
      return digital_root_rec(s)


print(digital_root_rec(65536))
print(digital_root_rec(942))
print(digital_root_rec(8))
```

**29.Напишите рекурсивную функцию для нахождения максимального элемента в
списке чисел.**

```python
def find_max_rec(my_list):
    """Найти максимальный элемент в списке 'numbers' рекурсивно."""
    if len(my_list) == 0:
      return my_list
    elif len(my_list) == 1:
      return my_list[0]
    else:
      mx = find_max_rec(my_list[1:])
      return my_list[0] if my_list[0] > mx else mx

my_list = [10, 45, 8, 92, 14, 55]
print(find_max_rec(my_list))
```
